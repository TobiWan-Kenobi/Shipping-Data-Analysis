```python
import pandas as pd
import os
import numpy as np
```

```python
# load saved data set
current_directory = os.path.abspath(".")
relative_path = "data/2019_AIS_assigend.csv"
full_file_path = os.path.join(current_directory, relative_path)
ais = pd.read_csv(full_file_path)
```

```python
ais
```

```python
# fill missing values with next available valid value
ais['next_nearest_port'] = ais.sort_values(by = 'TIME').groupby('IMO')['nearest_port'].fillna(method = 'bfill')
```

```python
# shift next nearest port up by 1 
ais['next_nearest_port'] = ais.sort_values(by = 'TIME').groupby('IMO')['next_nearest_port'].shift(-1)
```

```python
# do the same for country, longitude and latitude
ais['next_nearest_port_country'] = ais.sort_values(by = 'TIME').groupby('IMO')['nearest_port_country'].fillna(method = 'ffill')
ais['next_nearest_port_country'] = ais.sort_values(by = 'TIME').groupby('IMO')['next_nearest_port_country'].shift(-1)

ais['next_nearest_port_lon'] = ais.sort_values(by = 'TIME').groupby('IMO')['nearest_port_lon'].fillna(method = 'bfill')
ais['next_nearest_port_lon'] = ais.sort_values(by = 'TIME').groupby('IMO')['next_nearest_port_lon'].shift(-1)

ais['next_nearest_port_lat'] = ais.sort_values(by = 'TIME').groupby('IMO')['nearest_port_lat'].fillna(method = 'bfill')
ais['next_nearest_port_lat'] = ais.sort_values(by = 'TIME').groupby('IMO')['next_nearest_port_lat'].shift(-1)

```

```python
ais
```

```python
# since I am building a journey log, I will rename nearest_port now as start_port etc
ais.rename(columns={'nearest_port': 'start_port', 'nearest_port_country' : 'start_country', 'nearest_port_lon' : 'start_lon', 'nearest_port_lat' : 'start_lat', 'next_nearest_port': 'end_port', 'next_nearest_port_country' : 'end_country', 'next_nearest_port_lon' : 'end_lon', 'next_nearest_port_lat' : 'end_lat'}, inplace=True)
```

```python
# now feed forward start port, country, lat and lon
ais['start_port'] = ais.sort_values(by = 'TIME').groupby('IMO')['start_port'].fillna(method = 'ffill')
ais['start_country'] = ais.sort_values(by = 'TIME').groupby('IMO')['start_country'].fillna(method = 'ffill')
ais['start_lon'] = ais.sort_values(by = 'TIME').groupby('IMO')['start_lon'].fillna(method = 'ffill')
ais['start_lat'] = ais.sort_values(by = 'TIME').groupby('IMO')['start_lat'].fillna(method = 'ffill')
```

```python
ais['route'] = ais.start_port + ' - ' + ais.end_port
```

```python
# feed route forward
ais['route'] = ais.sort_values(by = 'TIME').groupby('IMO')['route'].fillna(method = 'ffill')
```

```python
# drop all columns that dont have a route - not necessary for the exercises here
ais = ais[ais['route'].notna()]
```

```python
ais.sort_values(['IMO', 'TIME'], inplace=True)
# Create an incremental count column within each group
ais['incremental_count'] = ais.groupby('IMO')['route'].transform(lambda x: x.ne(x.shift()).cumsum() - 1)
```

```python
ais_journey_log = ais.groupby(['IMO', 'start_port', 'start_country', 'start_lon', 'start_lat', 'end_port', 'end_country', 'end_lon', 'end_lat', 'route', 'incremental_count']).agg({
    'E_CO2_kg': 'sum',
    'TIME'    : ['min', 'max'],
    'SOG'     : 'mean'
}).reset_index()
```
